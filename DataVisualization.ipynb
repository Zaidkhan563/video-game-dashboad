```python
plt.figure(figsize=(12,5))
plt.bar(x=df["Genre"].value_counts().index , height=df["Genre"].value_counts().values)
plt.xticks(rotation=45 , fontsize=15)

plt.yticks(fontsize=12)

plt.show()
```

```python
# by using seaborn
sns.countplot(data=df , x="Genre")
```

```python
plt.figure(figsize=(12,5))
sns.countplot(data=df , x="Genre")
plt.xticks(rotation=45 , fontsize=15)
plt.yticks(fontsize=12)
plt.show()
```

```python
plt.figure(figsize=(12,5))
sns.countplot(data=df , x="Genre" , order=df["Genre"].value_counts().index)
plt.xticks(rotation=45 , fontsize=15)
plt.yticks(fontsize=12)
plt.show()
```

```python
plt.figure(figsize=(12,5))
sns.countplot(data=df , x="Genre" , order=df["Genre"].value_counts().index, width=0.2,color="grey")
plt.xticks(rotation=45 , fontsize=15)
plt.yticks(fontsize=12)
plt.show()
```
